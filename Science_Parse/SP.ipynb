```python
def get_GT(AN='135788877'):
    GT={}
    found = False
    with open(MFS_file,'r',encoding="utf8", errors='ignore') as tsvin:
        tsvin = csv.reader(tsvin, delimiter='\t')
        next(tsvin)
        for row in tsvin:
            if(row[0]!=AN):
                continue
            found=True
            #print(row)
            GT['TL']=row[2]
            GT['AN']=AN
            GT['abs']=row[20]
            GT['abs_code']=row[19]
            GT['title']=row[1]
            GT['auth'] = row[22]
            GT['email'] = row[24]
            GT ['doi'] = row [7]
            GT['inst']= row[26]
            GT['fp']=row[3]
            GT['lp']=row[4]
    return (found,GT)
```
